```python
%%writefile app.py
import os
import pickle
import streamlit as st    #######Streamlit  APP Code

from langchain_community.document_loaders import UnstructuredURLLoader, CSVLoader
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain_openai import OpenAIEmbeddings, ChatOpenAI
from langchain_community.vectorstores import FAISS
from langchain.chains import RetrievalQAWithSourcesChain

# ---------------- UI ----------------
st.set_page_config(page_title="News Research Tool", layout="wide")
st.title("ðŸ“° News Research Tool")

# ---------------- Sidebar ----------------
st.sidebar.header("News Article URLs")

urls = [
    st.sidebar.text_input("URL 1"),
    st.sidebar.text_input("URL 2"),
    st.sidebar.text_input("URL 3")
]

process_url_clicked = st.sidebar.button("Process URLs")

file_path = "faiss_store.pkl"

# ---------------- Main Area ----------------
main_placeholder = st.empty()

# ---------------- URL Processing ----------------
if process_url_clicked:
    if not any(urls):
        st.error("Please enter at least one URL")
    else:
        main_placeholder.text("Loading articles...")

        loader = UnstructuredURLLoader(urls=[u for u in urls if u])
        documents = loader.load()

        main_placeholder.text("Splitting text into chunks...")

        splitter = RecursiveCharacterTextSplitter(
            chunk_size=1000,
            chunk_overlap=200
        )
        docs = splitter.split_documents(documents)

        main_placeholder.text("Creating embeddings & FAISS index...")

        embeddings = OpenAIEmbeddings()
        vectorstore = FAISS.from_documents(docs, embeddings)

        with open(file_path, "wb") as f:
            pickle.dump(vectorstore, f)

        main_placeholder.success("Vector database created successfully!")

# ---------------- Question Answering ----------------
query = st.text_input("Ask a question about the articles:")

if query:
    if not os.path.exists(file_path):
        st.error("Please process URLs first")
    else:
        with open(file_path, "rb") as f:
            vectorstore = pickle.load(f)

        llm = ChatOpenAI(
            temperature=0,
            model_name="gpt-3.5-turbo"
        )

        chain = RetrievalQAWithSourcesChain.from_chain_type(
            llm=llm,
            chain_type="map_reduce",
            retriever=vectorstore.as_retriever(search_kwargs={"k": 4})
        )

        result = chain({"question": query})

        st.subheader("ðŸ“Œ Answer")
        st.write(result["answer"])

        st.subheader("ðŸ”— Sources")
        st.write(result["sources"])
```
